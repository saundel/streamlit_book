```python
import matplotlib.pyplot as plt
from palmerpenguins import load_penguins
import pandas as pd
```

```python
penguins = load_penguins()
penguins.head()
```

```python
fig, ax = plt.subplots(figsize=(15,10))
fig = plt.scatter(x=penguins["body_mass_g"], y=penguins["bill_length_mm"])
fig = plt.title("Penguin Body Mass (g) versus Bill Length (mm)")
fig = ax.set_xlabel("Body Mass (g)")
fig = ax.set_ylabel("Bill Length (mm)")
```

```python
import seaborn as sns
```

# Dashboard calcs

```python
penguins.head()
```

```python
yearly_specimen_counts = pd.DataFrame(penguins.groupby('species')['year'].value_counts()).reset_index()
yearly_specimen_counts
```

```python
year = max(penguins['year'])
prev_year = year - 1

prev_year_counts = yearly_specimen_counts[yearly_specimen_counts['year'] == prev_year]
prev_year_counts
```

```python
year_counts = yearly_specimen_counts[yearly_specimen_counts['year'] == year]
year_counts
```
